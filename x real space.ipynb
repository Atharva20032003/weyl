```python
h= np.zeros((2,2),dtype=np.complex_)
V= np.zeros((2,2),dtype=np.complex_)
```

```python
#parametres
tx=1
ty=1
tz=1
gamma=0.5
a=1
Nx=100
```

```python
def f(ky,kz):
    h[0][0]= tz*(2-np.cos(ky*a)+gamma-np.cos(kz*a))
    h[0][1]= (-1*ty*np.sin(ky*a))*1j
    h[1][0]=(ty*np.sin(ky*a))*1j
    h[1][1]=-tz*(2-np.cos(ky*a)+gamma-np.cos(kz*a))
    V[0][0]= -0.5*tz
    V[0][1]=(0.5*tx)*1j
    V[1][0]=(0.5*tx)*1j
    V[1][1]=(0.5*tz)
    Vdag=V.conj().T
    H=np.zeros((2*Nx,2*Nx),dtype=np.complex_)
    
    for k in range (Nx): #making H
        
        H[2*k][2*k]=h[0][0]
        H[2*k][2*k+1]=h[0][1]
        H[2*k+1][2*k]=h[1][0]
        H[2*k+1][2*k+1]=h[1][1]
        
        if (k!=(Nx-1)):
            H[2*k][2*k+2]=V[0][0]
            H[2*k][2*k+3]=V[0][1]
            H[2*k+1][2*k+2]=V[1][0]
            H[2*k+1][2*k+3]=V[1][1]
            
        if (k!=0):
            H[2*k][2*k-1]=Vdag[0][1]
            H[2*k][2*k-2]=Vdag[0][0]
            H[2*k+1][2*k-1]=Vdag[1][1]
            H[2*k+1][2*k-2]=Vdag[1][0]
        
    evalue, evect = np.linalg.eig(H)
    return evalue.real
        
 
```

```python
def g(ky,kz):
    h[0][0]= tz*(2-np.cos(ky*a)+gamma-np.cos(kz*a))
    h[0][1]= (-1*ty*np.sin(ky*a))*1j
    h[1][0]=(ty*np.sin(ky*a))*1j
    h[1][1]=-tz*(2-np.cos(ky*a)+gamma-np.cos(kz*a))
    V[0][0]= -0.5*tz
    V[0][1]=(0.5*tx)*1j
    V[1][0]=(0.5*tx)*1j
    V[1][1]=(0.5*tz)
    Vdag=V.conj().T
    H=np.zeros((2*Nx,2*Nx),dtype=np.complex_)
    
    for k in range (Nx): #making H
        
        H[2*k][2*k]=h[0][0]
        H[2*k][2*k+1]=h[0][1]
        H[2*k+1][2*k]=h[1][0]
        H[2*k+1][2*k+1]=h[1][1]
        
        if (k!=(Nx-1)):
            H[2*k][2*k+2]=V[0][0]
            H[2*k][2*k+3]=V[0][1]
            H[2*k+1][2*k+2]=V[1][0]
            H[2*k+1][2*k+3]=V[1][1]
            
        if (k!=0):
            H[2*k][2*k-1]=Vdag[0][1]
            H[2*k][2*k-2]=Vdag[0][0]
            H[2*k+1][2*k-1]=Vdag[1][1]
            H[2*k+1][2*k-2]=Vdag[1][0]
        
    evalue, evect = np.linalg.eig(H)
    return evect
```

```python
kzo= -0.905
for i in range (2*Nx):
    if np.abs(f(0,kzo)[i])<10**(-6):
        vecindex=i
        break
eigvec=g(0,kzo)[:,vecindex]

for i in range (2*Nx):
    if np.abs(f(0,kzo)[i])<10**(-6):
        vecindex1=i
        
eigvec1=g(0,kzo)[:,vecindex1]

c= (eigvec[2*Nx-2]/eigvec1[2*Nx-2]+eigvec[2*Nx-1]/eigvec1[2*Nx-1])/2

eigvecnew= eigvec- c*eigvec1

def depthf(n):
    return np.abs(eigvecnew[2*n])**2+np.abs(eigvecnew[2*n+1])**2
depth=np.zeros(Nx) 
intg=np.zeros(Nx)
for i in range (Nx):
  depth[i]= depthf(i)
for i in range (Nx):
  intg[i]=i
    
plt.plot(intg,depth)    
del vecindex
del vecindex1
    
```
